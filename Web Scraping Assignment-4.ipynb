# 1. Scrape the details of most viewed videos on YouTube from Wikipedia: Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos/ You need to find following details:
A) Rank

B) Name

C) Artist

D) Upload date

E) Views

```python
import selenium
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
import pandas as pd
import time
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
```

```python
rank=[]
name_of_video=[]
artist=[]
upload_date=[]
views=[]
```

```python
def youtube(url):
    driver=webdriver.Chrome('chromedriver.exe')
    driver.get(url)
    Rank=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']/tbody/tr/td[1]")
    for r in Rank[0:30]:
        rank.append(r.text)
        time.sleep(2)
    
    name=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']/tbody/tr/td[2]")
    for n in name[0:30]:
        name_of_video.append(n.text)
        time.sleep(2)
        
    art=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']/tbody/tr/td[3]")
    for a in art[0:30]:
        artist.append(a.text)
        time.sleep(2)
    
    Views=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']/tbody/tr/td[4]")
    for vw in Views[0:30]:
        views.append(vw.text)
        time.sleep(2)
        
    upload=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']/tbody/tr/td[5]")
    for up in upload[0:30]:
        upload_date.append(up.text)
        time.sleep(2)

    
    
    
```

```python
youtube('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
```

```python
top_video=pd.DataFrame({"Rank":rank[0:30],
                       "Video Name":name_of_video[0:30],
                       "Artist":artist[0:30],
                       "Views":views[0:30],
                       "Upload Date":upload_date[0:30]})
top_video
top_video.set_index('Rank')
```

```python
Series=[]
Place=[]
Date=[]
Month=[]
Time=[]
Match_Title=[]
```

```python
def Fixture(url):
    driver=webdriver.Chrome('chromedriver.exe')
    driver.get(url)
    fix=driver.find_element_by_xpath("//div[@class='navigation__menu-container']/div[2]/nav/ul/li[1]/div[2]/div/ul/li[1]/a")
    fixture=fix.get_attribute('href')
    driver.get(fixture)
    
    date=driver.find_elements_by_xpath("//div[@class='fixture__full-date']/span")
    for d in date:
        Date.append(d.text)
    
    month=driver.find_elements_by_xpath("//div[@class='fixture__date-details']/span[1]")
    for m in month:
        Month.append(m.text)
        
    time=driver.find_elements_by_xpath("//div[@class='fixture__date-details']/span[2]")
    for t in time:
        Time.append(t.text)
    
    
    place=driver.find_elements_by_xpath("//p[@class='fixture__additional-info']/span")
    for p in place:
        Place.append(p.text)
    
    
    series=driver.find_elements_by_xpath("//div[@class='fixture__format-strip']/span[2]")
    for s in series:
        Series.append(s.text)
    
    
    match_title=driver.find_elements_by_xpath("//div[@class='fixture__description u-unskewed-text']/p/strong")
    for mt in match_title:
        Match_Title.append(mt.text)
    
    

Fixture('https://www.bcci.tv/')    
```

# 5. Scrape the details of trending repositories on Github.com. Url = https://github.com/. You have to find the following details:
A) Repository title

B) Repository description

C) Contributors count

D) Language used

Note: - From the home page you have to click on the trending option from Explore menu through code.

```python
import pandas as pd
import selenium
import time
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
```

```python
Repository_Name=[]
Repository_Description=[]
Contributer_Count=[]
Language_used=[]
```

```python
def github(url):
    driver=webdriver.Chrome('chromedriver.exe')
    driver.get(url)
    
    trending=driver.find_element_by_xpath("//nav[@class='mt-0 px-3 px-lg-0 mb-5 mb-lg-0']/ul/li[4]/details/summary")
    trending.click()
    
    trending1=driver.find_element_by_xpath("//div[@class='dropdown-menu flex-auto rounded px-0 pt-2 pb-0 mt-0 pb-4 p-lg-4 position-relative position-lg-absolute left-0 left-lg-n4']/ul[2]/li[3]/a")
    trending2=trending1.get_attribute('href')
    driver.get(trending2)
    
    time.sleep(2)
    try:
        rep_nam=driver.find_elements_by_xpath("//article[@class='Box-row']/h1/a")
        for rn in rep_nam:
            Repository_Name.append(rn.text)
    except NoSuchElementException as e:
        Repository_Name.append('-')
    
    time.sleep(2)
    try:
        rep_des=driver.find_elements_by_xpath("//p[@class='col-9 color-text-secondary my-1 pr-4']")
        for rd in rep_des:
            Repository_Description.append(rd.text)
    except NoSuchElementException as e:
        Repository_Description.append('-')
    
    time.sleep(2)
    try:
        contri_count=driver.find_elements_by_xpath("//div[@class='f6 color-text-secondary mt-2']/a[2]")
        for cc in contri_count:
            Contributer_Count.append(cc.text)
    except NoSuchElementException as e:
        Contributer_Count.append('-')
        
    time.sleep(2)
    try:
        lang_used=driver.find_elements_by_xpath("//div[@class='f6 color-text-secondary mt-2']/span[1]/span[2]")
        for lu in lang_used:
            Language_used.append(lu.text)
    except NoSuchElementException as e:
        Language_used.append('-')
    
    
github('https://github.com/')

    
```

```python
GitHub=pd.DataFrame({'Repository title':Repository_Name[0:20],'Repository description':Repository_Description[0:20],'Contributors count':Contributer_Count[0:20],'Language used':Language_used[0:20]})
GitHub
```

```python
import pandas as pd
import time
import selenium
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
```

```python
Name=[]
Designation=[]
Company=[]
Skills_they_Hire=[]
Location=[]
```

```python
def naukri(url):
    driver=webdriver.Chrome('chromedriver.exe')
    driver.get(url)
    
    recruiter=driver.find_element_by_xpath("//div[@class='subMenu']/ul/li[1]/a")
    url=recruiter.get_attribute('href')
    driver.get(url)
    
    search=driver.find_element_by_xpath("//div[@id='skill']/div[1]/div[2]/input")
    search.send_keys('Data Science')
    search_button=driver.find_element_by_xpath("//button[@id='qsbFormBtn']").click()
    
    for page in range(0,116):
        
        name=driver.find_elements_by_xpath("//a[@class='ellipsis']/span")
        for n in name:
            Name.append(n.text)
        
        designation=driver.find_elements_by_xpath("//p[@class='highlightable']/span[1]")
        for d in designation:
            Designation.append(d.text)

        company=driver.find_elements_by_xpath("//p[@class='highlightable']/a[2]")
        for c in company:
            Company.append(c.text)

        skill=driver.find_elements_by_xpath("//div[@class='hireSec highlightable']")
        for s in skill:
            Skills_they_Hire.append(s.text)
    
        location=driver.find_elements_by_xpath("//p[@class='highlightable']/span[2]/small")
        for l in location:
            Location.append(l.text)

        next_button=driver.find_elements_by_xpath("//div[@class='pagination']/button")
        for i in next_button:
            if i.text=='Next':
                i.click()
                
            
            
naukri('https://www.naukri.com/')
```
